## Best Advice Premium Loop

This script cycles through a .csv file that specifies policy details, and outputs the premiums for all Irish life insurance providers

The script requires Selenium and ChromeDriver. Please be sure to read all comments before running.

Run the below cell to initialise all the required packages.

```python
import pandas as pd
from time import sleep
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait


# bestadvice credentials

# initialize the Chrome driver
# this needs to be downloaded from the below link
#
#
# https://chromedriver.chromium.org/
#
#
# and placed in the working directory too


driver = webdriver.Chrome("/Users/davkav/chromedriver") # link the working directory here

# head to bestadvice login page
driver.get("https://www.bis-platform.com/ba/login.php?redirect=/ba/b_term.php")
```

```python
# you need to create and place a blank premiums sheet into working directory
df = pd.read_csv('premiums_single.csv')


# run the loop for single life premiums
for i in df.index:
#for i in range(60,76):
    entry = df.loc[i]

    # enter the details in row 0
    if entry.smoker == 0:
        smoker = "No"
    else:
        smoker = "Yes"
    age = int(entry.age)
    term = int(entry.term)
    sumass = int(entry.sumass)
    driver.find_element_by_id("age1").clear()
    driver.find_element_by_id("age1").send_keys(age)
    driver.find_element_by_name("term").clear()
    driver.find_element_by_name("term").send_keys(term)
    driver.find_element_by_name("lc1").clear()
    driver.find_element_by_name("lc1").send_keys(sumass)
    driver.find_element_by_name("ci1").clear()
    driver.find_element_by_name("ci1").send_keys(sumass)
    
    ddsmoker = Select(driver.find_element_by_name("smoker1"))
    ddsmoker.select_by_visible_text(smoker)
    
    #run the quotation
    driver.find_element_by_xpath('//*[@id="frmTerm"]/table/tbody/tr[24]/td/div').click()

    #take Royal London premiums
    prem1_LC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[3]/td[2]').text+ '0'
    df.at[i,'prem1_LC']=prem1_LC.replace(',','')
    
    prem1_AC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[3]/td[2]').text+ '0'
    df.at[i,'prem1_AC']=prem1_AC.replace(',','')
    
    prem1_SA = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[3]/tbody/tr[3]/td[2]').text+ '0'
    df.at[i,'prem1_SA']=prem1_SA.replace(',','')
    
    prem1_CI = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[4]/tbody/tr[3]/td[2]').text+ '0'
    df.at[i,'prem1_CI']=prem1_CI.replace(',','')
    
    prem1_ML = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[3]/td[6]').text+ '0'
    df.at[i,'prem1_ML']=prem1_ML.replace(',','')
    
    prem1_MC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[3]/td[6]').text+ '0'
    df.at[i,'prem1_MC']=prem1_MC.replace(',','')
    
    #take Zurich premiums
    prem2_LC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[4]/td[2]').text+ '0'
    df.at[i,'prem2_LC']=prem2_LC.replace(',','')
    
    prem2_AC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[4]/td[2]').text+ '0'
    df.at[i,'prem2_AC']=prem2_AC.replace(',','')
    
    prem2_SA = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[3]/tbody/tr[4]/td[2]').text+ '0'
    df.at[i,'prem2_SA']=prem2_SA.replace(',','')
    
    prem2_CI = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[4]/tbody/tr[4]/td[2]').text+ '0'
    df.at[i,'prem2_CI']=prem2_CI.replace(',','')
    
    prem2_ML = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[4]/td[6]').text+ '0'
    df.at[i,'prem2_ML']=prem2_ML.replace(',','')
    
    prem2_MC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[4]/td[6]').text+ '0'
    df.at[i,'prem2_MC']=prem2_MC.replace(',','')
    
    # take Aviva premiums
    prem3_LC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[5]/td[2]').text+ '0'
    df.at[i,'prem3_LC']=prem3_LC.replace(',','')
    
    prem3_AC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[5]/td[2]').text+ '0'
    df.at[i,'prem3_AC']=prem3_AC.replace(',','')
    
    prem3_SA = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[3]/tbody/tr[5]/td[2]').text+ '0'
    df.at[i,'prem3_SA']=prem3_SA.replace(',','')
    
    prem3_CI = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[4]/tbody/tr[5]/td[2]').text+ '0'
    df.at[i,'prem3_CI']=prem3_CI.replace(',','')
    
    prem3_ML = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[5]/td[6]').text+ '0'
    df.at[i,'prem3_ML']=prem3_ML.replace(',','')
    
    prem3_MC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[5]/td[6]').text+ '0'
    df.at[i,'prem3_MC']=prem3_MC.replace(',','')
    
    # take Irish Life premiums
    prem4_LC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[6]/td[2]').text+ '0'
    df.at[i,'prem4_LC']=prem4_LC.replace(',','')
    
    prem4_AC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[6]/td[2]').text+ '0'
    df.at[i,'prem4_AC']=prem4_AC.replace(',','')
    
    prem4_SA = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[3]/tbody/tr[6]/td[2]').text+ '0'
    df.at[i,'prem4_SA']=prem4_SA.replace(',','')
    
    prem4_CI = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[4]/tbody/tr[6]/td[2]').text+ '0'
    df.at[i,'prem4_CI']=prem4_CI.replace(',','')
    
    prem4_ML = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[6]/td[6]').text+ '0'
    df.at[i,'prem4_ML']=prem4_ML.replace(',','')
    
    prem4_MC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[6]/td[6]').text+ '0'
    df.at[i,'prem4_MC']=prem4_MC.replace(',','')
    
    # take New Ireland premiums
    prem5_LC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[7]/td[2]').text+ '0'
    df.at[i,'prem5_LC']=prem5_LC.replace(',','')
    
    prem5_AC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[7]/td[2]').text+ '0'
    df.at[i,'prem5_AC']=prem5_AC.replace(',','')
    
    prem5_SA = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[3]/tbody/tr[7]/td[2]').text+ '0'
    df.at[i,'prem5_SA']=prem5_SA.replace(',','')
    
    prem5_CI = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[4]/tbody/tr[7]/td[2]').text+ '0'
    df.at[i,'prem5_CI']=prem5_CI.replace(',','')
    
    prem5_ML = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[1]/tbody/tr[7]/td[6]').text+ '0'
    df.at[i,'prem5_ML']=prem5_ML.replace(',','')
    
    prem5_MC = driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[1]/table[2]/tbody/tr[7]/td[6]').text+ '0'
    df.at[i,'prem5_MC']=prem5_MC.replace(',','')
    
    driver.find_element_by_xpath('/html/body/table/tbody/tr[4]/td[2]/div[1]/div[1]/div').click()
    
df.to_csv('premiums_filled_single.csv')
```
